```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv('EEPledge.csv', dtype=str)
```

```python
data
```

```python
dataf = pd.DataFrame(data = data)
```

```python
dataf_country = dataf.groupby(['Country']).size()
dataf_country.plot(kind='bar')
plt.show()
```

```python
dataf_age = dataf.groupby(['Age']).size()
dataf_gender = dataf.groupby(['Gender']).size()
dataf_date = dataf.groupby(['Date']).size()
```

```python
dataf_country.sort_values(ascending = False).head(10).plot(kind='bar')
plt.show()
```

```python
dataf_age.sort_values(ascending = False).head().plot(kind='bar')
plt.show()
```

```python
dataf_date.sort_values(ascending = False).head().plot(kind='bar')
plt.show()
```

```python
dataf_gender.sort_values(ascending = False).head().plot(kind='pie')
plt.show()
```

```python
dataf_age[1:].sort_values(ascending = False).head(10).plot(kind='bar')
plt.show()
```
